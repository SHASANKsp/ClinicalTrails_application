```python
from neo4j import GraphDatabase

driver = GraphDatabase.driver("bolt://localhost:7687", auth=("neo4j", "CTrail@123"))
```

```python
def get_labels(tx):
    result = tx.run("CALL db.labels()")
    return [record["label"] for record in result]

with driver.session() as session:
    labels = session.read_transaction(get_labels)
    print(labels)
```

```python
def get_relationships(tx):
    result = tx.run("CALL db.relationshipTypes()")
    return [record["relationshipType"] for record in result]

with driver.session() as session:
    rels = session.read_transaction(get_relationships)
    print(rels)
```

```python
def get_node_properties(tx):
    query = """
    CALL db.schema.nodeTypeProperties()
    YIELD nodeLabels, propertyName
    RETURN nodeLabels, collect(propertyName) AS properties
    """
    result = tx.run(query)

    schema = {}
    for record in result:
        label = record["nodeLabels"][0]  # usually single label
        props = record["properties"]
        schema[label] = props

    return schema

with driver.session() as session:
    node_schema = session.read_transaction(get_node_properties)
    print(node_schema)
```

# Application

```python
from neo4j import GraphDatabase
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx

# Neo4j connection
class KGClient:
    def __init__(self, uri, user, password):
        self.driver = GraphDatabase.driver(uri, auth=(user, password))

    def query(self, cypher, params={}):
        with self.driver.session() as session:
            result = session.run(cypher, params)
            return [dict(r) for r in result]
```

```python
URI = "neo4j://localhost:7687" 
USER = "neo4j"
PASSWORD = "CTrail@123"

# Step 2: Initialize KG client
kg = KGClient(URI, USER, PASSWORD)
```

## Sponsor Intelligence Dashboard

```python
def sponsor_trial_counts(kg: KGClient):
    query = """
    MATCH (s:Sponsor)-[:SPONSORS]->(st:Study)
    RETURN s.name AS sponsor, COUNT(st) AS trial_count
    ORDER BY trial_count DESC
    """
    data = kg.query(query)
    df = pd.DataFrame(data)
    return df
```

```python
def plot_top_sponsors(df, top_n=10):
    df = df.head(top_n)

    plt.figure(figsize=(10,6))
    plt.barh(df['sponsor'], df['trial_count'])
    plt.xlabel("Number of Trials")
    plt.ylabel("Sponsor")
    plt.title("Top Sponsors by Trial Count")
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.savefig("top_sponsors.png", dpi=300)
    plt.show()
```

```python
df = sponsor_trial_counts(kg)
plot_top_sponsors(df)
```

## Geo Distribution of Trials

```python
def trials_by_country(kg):
    query = """
    MATCH (l:Location)<-[:CONDUCTED_AT]-(st:Study)
    RETURN l.country AS country, COUNT(st) AS trials
    ORDER BY trials DESC
    """
    return pd.DataFrame(kg.query(query))
```

```python
def plot_country_distribution(df):
    df = df.head(15)

    plt.figure(figsize=(10,6))
    plt.bar(df['country'], df['trials'])
    plt.xticks(rotation=45, ha='right')
    plt.ylabel("Trials")
    plt.title("Clinical Trials by Country")
    plt.tight_layout()
    plt.savefig("trials_by_country.png", dpi=300)
    plt.show()
```

```python
df = trials_by_country(kg)
plot_country_distribution(df)
```

## Intervention Landscape (Drug Popularity)

```python
def intervention_frequency(kg):
    query = """
    MATCH (i:Intervention)<-[:USES_INTERVENTION]-(st:Study)
    RETURN i.name AS intervention, COUNT(st) AS freq
    ORDER BY freq DESC
    """
    return pd.DataFrame(kg.query(query))
```

```python
def plot_top_interventions(df, top_n=10):
    df = df.head(top_n)

    plt.figure(figsize=(10,6))
    plt.barh(df['intervention'], df['freq'])
    plt.xlabel("Number of Trials")
    plt.title("Top Interventions")
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.savefig("top_interventions.png", dpi=300)
    plt.show()
```

```python
df = intervention_frequency(kg)
plot_top_interventions(df)
```

## arm

```python
def arm_type_distribution(kg):
    query = """
    MATCH (st:Study)-[:HAS_ARM]->(a:Arm)
    RETURN a.type AS arm_type, COUNT(a) AS count
    """
    return pd.DataFrame(kg.query(query))
```

```python
def plot_arm_distribution(df):
    plt.figure(figsize=(8,6))
    plt.pie(df['count'], labels=df['arm_type'], autopct='%1.1f%%')
    plt.title("Arm Type Distribution")
    plt.tight_layout()
    plt.savefig("arm_distribution.png", dpi=300)
    plt.show()
```

```python
df = arm_type_distribution(kg)
plot_arm_distribution(df)
```

## study phase distribution

```python
def study_phase_distribution(kg):
    query = """
    MATCH (st:Study)
    WHERE st.phases IS NOT NULL
    RETURN st.phases AS phase, COUNT(st) AS count
    ORDER BY count DESC
    """
    return pd.DataFrame(kg.query(query))
```

```python
def plot_phase_distribution(df):
    plt.figure(figsize=(15,6))
    plt.bar(df['phase'], df['count'])
    plt.xlabel("Phase")
    plt.ylabel("Count")
    plt.title("Study Phase Distribution")
    plt.tight_layout()
    plt.savefig("phase_distribution.png", dpi=300)
    plt.show()
```

```python
df = study_phase_distribution(kg)
plot_phase_distribution(df)
```

```python
def get_trials_by_drug(kg, drug_name):
    query = """
    MATCH (st:Study)-[:USES_INTERVENTION]->(i:Intervention)
    WHERE toLower(i.name) CONTAINS toLower($drug)
    RETURN st.nct_id AS trial,
           st.phases AS phase,
           st.enrollment AS enrollment
    """
    return pd.DataFrame(kg.query(query, {"drug": drug_name}))
```

```python
def plot_drug_phase_distribution(df, drug_name):
    df = df.groupby("phase").size().reset_index(name="count")

    plt.figure(figsize=(15,6))
    plt.bar(df["phase"], df["count"])
    plt.title(f"Phase Distribution for {drug_name}")
    plt.xlabel("Phase")
    plt.ylabel("Number of Trials")
    plt.tight_layout()
    plt.savefig(f"{drug_name}_phase.png", dpi=300)
    plt.show()
```

```python
df = get_trials_by_drug(kg, "Donepezil")
plot_drug_phase_distribution(df, "Donepezil")
```

```python
def get_interventions_by_condition(kg, condition):
    query = """
    MATCH (st:Study)-[:STUDIES]->(c:Condition)
    MATCH (st)-[:USES_INTERVENTION]->(i:Intervention)
    WHERE toLower(c.name) CONTAINS toLower($condition)
    RETURN i.name AS intervention, COUNT(st) AS trials
    ORDER BY trials DESC
    """
    return pd.DataFrame(kg.query(query, {"condition": condition}))
```

```python
def plot_condition_interventions(df, condition):
    df = df.head(10)

    plt.figure(figsize=(10,6))
    plt.barh(df["intervention"], df["trials"])
    plt.title(f"Top Interventions for {condition}")
    plt.tight_layout()
    plt.savefig(f"{condition}_interventions.png", dpi=300)
    plt.show()
```

```python
df = get_interventions_by_condition(kg, "Alzheimer")
plot_condition_interventions(df, "Alzheimer")
```

```python
def get_sponsor_portfolio(kg, sponsor):
    query = """
    MATCH (st:Study)<-[:SPONSORS]-(s:Sponsor)
    MATCH (st)-[:STUDIES]->(c:Condition)
    WHERE toLower(s.name) CONTAINS toLower($sponsor)
    RETURN c.name AS condition, COUNT(st) AS trials
    ORDER BY trials DESC
    """
    return pd.DataFrame(kg.query(query, {"sponsor": sponsor}))
```

```python
def plot_sponsor_focus(df, sponsor):
    df = df.head(10)

    plt.figure(figsize=(10,6))
    plt.barh(df["condition"], df["trials"])
    plt.title(f"{sponsor} - Condition Focus")
    plt.tight_layout()
    plt.savefig(f"{sponsor}_portfolio.png", dpi=300)
    plt.show()
```

```python
df = get_sponsor_portfolio(kg, "Pfizer")
plot_sponsor_focus(df, "Pfizer")
```

```python
def get_drug_subgraph(kg, drug):
    query = """
    MATCH (i:Intervention)<-[:USES_INTERVENTION]-(st:Study)
    MATCH (st)-[:STUDIES]->(c:Condition)
    MATCH (st)<-[:SPONSORS]-(s:Sponsor)
    WHERE toLower(i.name) CONTAINS toLower($drug)
    RETURN i.name AS drug, c.name AS condition, s.name AS sponsor
    """
    data = kg.query(query, {"drug": drug})

    G = nx.Graph()
    for row in data:
        G.add_edge(row["drug"], row["condition"])
        G.add_edge(row["drug"], row["sponsor"])

    return G
```

```python
def plot_subgraph(G, title):
    plt.figure(figsize=(10,8))
    pos = nx.spring_layout(G)

    nx.draw(G, pos, with_labels=True, node_size=200, font_size=8)

    plt.title(title)
    plt.tight_layout()
    plt.savefig("subgraph.png", dpi=300)
    plt.show()
```

```python
G = get_drug_subgraph(kg, "Donepezil")
plot_subgraph(G, "Drug-Centric Subgraph")
```

```python
def drug_evidence_bundle(kg, drug):
    query = """
    MATCH (i:Intervention)<-[:USES_INTERVENTION]-(st:Study)
    OPTIONAL MATCH (st)-[:STUDIES]->(c:Condition)
    OPTIONAL MATCH (st)<-[:SPONSORS]-(s:Sponsor)
    OPTIONAL MATCH (st)-[:CONDUCTED_AT]->(l:Location)
    WHERE toLower(i.name) CONTAINS toLower($drug)

    RETURN i.name AS drug,
           st.nct_id AS trial,
           st.phases AS phase,
           c.name AS condition,
           s.name AS sponsor,
           l.country AS country
    """

    data = kg.query(query, {"drug": drug})
    df = pd.DataFrame(data)

    # Aggregations
    agg = {
        "top_conditions": df["condition"].value_counts().head(5).to_dict(),
        "top_sponsors": df["sponsor"].value_counts().head(5).to_dict(),
        "phase_distribution": df["phase"].value_counts().to_dict()
    }

    return {
        "query": f"Drug analysis: {drug}",
        "evidence_chunks": data,
        "aggregations": agg
    }
```

```python
result = drug_evidence_bundle(kg, "Donepezil")
```
