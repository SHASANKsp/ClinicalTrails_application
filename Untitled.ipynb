```python
def safe_get(d, *keys):
    for key in keys:
        if isinstance(d, dict):
            d = d.get(key)
        else:
            return None
    return d


def safe_join(values):
    if not values:
        return None
    clean = [str(v) for v in values if v is not None]
    return ";".join(clean) if clean else None


def extract_numeric(value):
    if value is None:
        return None
    match = re.search(r"[-+]?\d*\.\d+|\d+", str(value))
    return float(match.group()) if match else None


# =========================
# Extraction Logic
# =========================

def extract_study(study):
    rows = []

    protocol = study.get("protocolSection", {})
    results = study.get("resultsSection", {})

    # ---- Identification ----
    nct_id = safe_get(protocol, "identificationModule", "nctId")
    brief_title = safe_get(protocol, "identificationModule", "briefTitle")
    official_title = safe_get(protocol, "identificationModule", "officialTitle")

    # ---- Status ----
    status = safe_get(protocol, "statusModule", "overallStatus")

    # ---- Design ----
    study_type = safe_get(protocol, "designModule", "studyType")
    phases = safe_get(protocol, "designModule", "phases")
    enrollment = safe_get(protocol, "designModule", "enrollmentInfo", "count")

    # ---- Conditions ----
    conditions = safe_get(protocol, "conditionsModule", "conditions")
    keywords = safe_get(protocol, "conditionsModule", "keywords")

    # ---- Sponsor ----
    lead_sponsor = safe_get(protocol, "sponsorCollaboratorsModule", "leadSponsor", "name")

    # ---- Eligibility ----
    minimum_age = safe_get(protocol, "eligibilityModule", "minimumAge")
    maximum_age = safe_get(protocol, "eligibilityModule", "maximumAge")
    sex = safe_get(protocol, "eligibilityModule", "sex")

    # ---- Results availability ----
    has_results = study.get("hasResults")

    # ---- Interventions ----
    interventions = safe_get(protocol, "armsInterventionsModule", "interventions") or []

    # ---- Primary Outcomes ----
    primary_outcomes = safe_get(protocol, "outcomesModule", "primaryOutcomes") or []

    outcome_titles = [
        o.get("measure")
        for o in primary_outcomes
        if o.get("measure") is not None
    ]

    outcome_timeframes = [
        o.get("timeFrame")
        for o in primary_outcomes
        if o.get("timeFrame") is not None
    ]

    # ---- Results Extraction ----
    outcome_values = []

    outcome_measures = safe_get(results, "outcomeMeasuresModule", "outcomeMeasures") or []

    for om in outcome_measures:
        classes = om.get("classes") or []
        for cls in classes:
            categories = cls.get("categories") or []
            for cat in categories:
                measurements = cat.get("measurements") or []
                for m in measurements:
                    numeric_val = extract_numeric(m.get("value"))
                    if numeric_val is not None:
                        outcome_values.append(numeric_val)

    mean_numeric = (
        sum(outcome_values) / len(outcome_values)
        if outcome_values else None
    )

    # ---- Flatten per intervention ----
    if not interventions:
        rows.append({
            "nct_id": nct_id,
            "brief_title": brief_title,
            "official_title": official_title,
            "status": status,
            "study_type": study_type,
            "phases": safe_join(phases),
            "enrollment": enrollment,
            "conditions": safe_join(conditions),
            "keywords": safe_join(keywords),
            "lead_sponsor": lead_sponsor,
            "minimum_age": minimum_age,
            "maximum_age": maximum_age,
            "sex": sex,
            "has_results": has_results,
            "intervention_type": None,
            "intervention_name": None,
            "primary_outcomes": safe_join(outcome_titles),
            "outcome_time_frames": safe_join(outcome_timeframes),
            "num_numeric_outcomes": len(outcome_values),
            "mean_numeric_outcome": mean_numeric
        })
    else:
        for intervention in interventions:
            rows.append({
                "nct_id": nct_id,
                "brief_title": brief_title,
                "official_title": official_title,
                "status": status,
                "study_type": study_type,
                "phases": safe_join(phases),
                "enrollment": enrollment,
                "conditions": safe_join(conditions),
                "keywords": safe_join(keywords),
                "lead_sponsor": lead_sponsor,
                "minimum_age": minimum_age,
                "maximum_age": maximum_age,
                "sex": sex,
                "has_results": has_results,
                "intervention_type": intervention.get("type"),
                "intervention_name": intervention.get("name"),
                "primary_outcomes": safe_join(outcome_titles),
                "outcome_time_frames": safe_join(outcome_timeframes),
                "num_numeric_outcomes": len(outcome_values),
                "mean_numeric_outcome": mean_numeric
            })

    return rows


# =========================
# Streaming Execution
# =========================

def run_extraction():
    file_size = os.path.getsize(INPUT_FILE)

    fieldnames = [
        "nct_id",
        "brief_title",
        "official_title",
        "status",
        "study_type",
        "phases",
        "enrollment",
        "conditions",
        "keywords",
        "lead_sponsor",
        "minimum_age",
        "maximum_age",
        "sex",
        "has_results",
        "intervention_type",
        "intervention_name",
        "primary_outcomes",
        "outcome_time_frames",
        "num_numeric_outcomes",
        "mean_numeric_outcome"
    ]

    total_studies = 0

    with open(OUTPUT_FILE, "w", newline="", encoding="utf-8") as out_csv:
        writer = csv.DictWriter(out_csv, fieldnames=fieldnames)
        writer.writeheader()

        with open(INPUT_FILE, "rb") as f:
            parser = ijson.items(f, "item")

            with tqdm(total=file_size,
                      unit="B",
                      unit_scale=True,
                      desc="Processing ClinicalTrials") as pbar:
                
                for study in parser:
                    extracted_rows = extract_study(study)

                    for row in extracted_rows:
                        writer.writerow(row)

                    total_studies += 1
                    
                    pbar.update(f.tell() - pbar.n)
                    pbar.set_postfix(studies=total_studies)

    print("\nExtraction complete.")
    print("Total studies processed:", total_studies)
    print("Output saved to:", OUTPUT_FILE)


# =========================
# Run
# =========================

if __name__ == "__main__":
    run_extraction()
```

```python
import ijson
import os
import csv
from tqdm import tqdm


INPUT_FILE = "ctg-studies.json"
OUTPUT_FILE = "clinical_trials_master_full_extract.csv"


def safe_get(d, *keys):
    for key in keys:
        if isinstance(d, dict):
            d = d.get(key)
        else:
            return None
    return d


def safe_join(values):
    if not values:
        return None
    clean = [str(v) for v in values if v is not None]
    return ";".join(clean) if clean else None


def extract_study(study):

    protocol = study.get("protocolSection", {})
    results = study.get("resultsSection", {})
    derived = study.get("derivedSection", {})

    row = {}

    # ---------------------------
    # Identification
    # ---------------------------
    idmod = protocol.get("identificationModule", {})
    row["nct_id"] = idmod.get("nctId")
    row["brief_title"] = idmod.get("briefTitle")
    row["official_title"] = idmod.get("officialTitle")
    row["acronym"] = idmod.get("acronym")
    row["organization"] = safe_get(idmod, "organization", "fullName")
    row["organization_class"] = safe_get(idmod, "organization", "class")

    # ---------------------------
    # Status & Dates
    # ---------------------------
    status = protocol.get("statusModule", {})
    row["overall_status"] = status.get("overallStatus")
    row["start_date"] = safe_get(status, "startDateStruct", "date")
    row["primary_completion_date"] = safe_get(status, "primaryCompletionDateStruct", "date")
    row["completion_date"] = safe_get(status, "completionDateStruct", "date")
    row["first_submit_date"] = status.get("studyFirstSubmitDate")
    row["first_post_date"] = safe_get(status, "studyFirstPostDateStruct", "date")
    row["last_update_post_date"] = safe_get(status, "lastUpdatePostDateStruct", "date")
    row["has_results"] = study.get("hasResults")

    # ---------------------------
    # Design
    # ---------------------------
    design = protocol.get("designModule", {})
    row["study_type"] = design.get("studyType")
    row["phases"] = safe_join(design.get("phases"))
    row["enrollment"] = safe_get(design, "enrollmentInfo", "count")
    row["enrollment_type"] = safe_get(design, "enrollmentInfo", "type")
    row["allocation"] = safe_get(design, "designInfo", "allocation")
    row["intervention_model"] = safe_get(design, "designInfo", "interventionModel")
    row["primary_purpose"] = safe_get(design, "designInfo", "primaryPurpose")
    row["masking"] = safe_get(design, "designInfo", "maskingInfo", "masking")
    row["has_dmc"] = safe_get(protocol, "oversightModule", "oversightHasDmc")

    # ---------------------------
    # Sponsor
    # ---------------------------
    sponsor = protocol.get("sponsorCollaboratorsModule", {})
    row["lead_sponsor"] = safe_get(sponsor, "leadSponsor", "name")
    row["lead_sponsor_class"] = safe_get(sponsor, "leadSponsor", "class")

    # ---------------------------
    # Conditions
    # ---------------------------
    row["conditions"] = safe_join(
        safe_get(protocol, "conditionsModule", "conditions")
    )
    row["keywords"] = safe_join(
        safe_get(protocol, "conditionsModule", "keywords")
    )

    row["mesh_conditions"] = safe_join([
        m.get("term")
        for m in safe_get(derived, "conditionBrowseModule", "meshes") or []
    ])

    # ---------------------------
    # Interventions
    # ---------------------------
    interventions = safe_get(protocol, "armsInterventionsModule", "interventions") or []
    row["intervention_names"] = safe_join([i.get("name") for i in interventions])
    row["intervention_types"] = safe_join([i.get("type") for i in interventions])

    row["mesh_interventions"] = safe_join([
        m.get("term")
        for m in safe_get(derived, "interventionBrowseModule", "meshes") or []
    ])

    # ---------------------------
    # Outcomes
    # ---------------------------
    primary_outcomes = safe_get(protocol, "outcomesModule", "primaryOutcomes") or []
    row["primary_outcome_measures"] = safe_join([o.get("measure") for o in primary_outcomes])
    row["primary_outcome_timeframes"] = safe_join([o.get("timeFrame") for o in primary_outcomes])

    secondary_outcomes = safe_get(protocol, "outcomesModule", "secondaryOutcomes") or []
    row["secondary_outcome_measures"] = safe_join([o.get("measure") for o in secondary_outcomes])

    # ---------------------------
    # Eligibility
    # ---------------------------
    row["minimum_age"] = safe_get(protocol, "eligibilityModule", "minimumAge")
    row["maximum_age"] = safe_get(protocol, "eligibilityModule", "maximumAge")
    row["sex"] = safe_get(protocol, "eligibilityModule", "sex")
    row["healthy_volunteers"] = safe_get(protocol, "eligibilityModule", "healthyVolunteers")
    row["std_ages"] = safe_join(safe_get(protocol, "eligibilityModule", "stdAges"))
    row["eligibility_text"] = safe_get(protocol, "eligibilityModule", "eligibilityCriteria")

    # ---------------------------
    # Geography
    # ---------------------------
    locations = safe_get(protocol, "contactsLocationsModule", "locations") or []
    row["countries"] = safe_join([l.get("country") for l in locations])
    row["cities"] = safe_join([l.get("city") for l in locations])

    return row


def run_extraction():

    file_size = os.path.getsize(INPUT_FILE)
    total = 0

    with open(OUTPUT_FILE, "w", newline="", encoding="utf-8") as out_csv:

        writer = None

        with open(INPUT_FILE, "rb") as f:
            parser = ijson.items(f, "item")

            with tqdm(total=file_size,
                      unit="B",
                      unit_scale=True,
                      desc="Extracting Master Dataset") as pbar:

                for study in parser:
                    row = extract_study(study)

                    if writer is None:
                        writer = csv.DictWriter(out_csv, fieldnames=row.keys())
                        writer.writeheader()

                    writer.writerow(row)
                    total += 1
                    pbar.update(f.tell() - pbar.n)
                    pbar.set_postfix(studies=total)

    print("\nExtraction complete.")
    print("Total studies:", total)


if __name__ == "__main__":
    run_extraction()
```

```python
def trial_completion_rate(df):
    summary = (
        df.groupby("overall_status")
          .size()
          .reset_index(name="count")
    )
    summary["percentage"] = summary["count"] / summary["count"].sum() * 100
    return summary
def results_reporting_rate_by_intervention(df):
    temp = df.explode("intervention_types".split(";"))
    summary = (
        df.groupby("intervention_types")
          .agg(
              total_trials=("nct_id", "count"),
              results_posted=("has_results", "sum")
          )
    )
    summary["reporting_rate"] = summary["results_posted"] / summary["total_trials"]
    return summary
def intervention_condition_matrix(df):
    exploded = df.assign(
        intervention=df["intervention_names"].str.split(";"),
        condition=df["conditions"].str.split(";")
    ).explode("intervention").explode("condition")

    return pd.crosstab(exploded["intervention"], exploded["condition"])
def trial_duration_analysis(df):
    df2 = df.copy()
    df2["start_date"] = pd.to_datetime(df2["start_date"], errors="coerce")
    df2["completion_date"] = pd.to_datetime(df2["completion_date"], errors="coerce")
    df2["duration_days"] = (df2["completion_date"] - df2["start_date"]).dt.days

    return df2.groupby("phases")["duration_days"].describe()
def geographic_distribution(df):
    exploded = df.assign(country=df["countries"].str.split(";")).explode("country")
    return exploded["country"].value_counts()
def eligibility_length_analysis(df):
    df["eligibility_length"] = df["eligibility_text"].fillna("").str.len()
    return df.groupby("conditions")["eligibility_length"].mean()
def sponsor_success_rate(df):
    summary = (
        df.groupby("lead_sponsor_class")
          .agg(
              total=("nct_id", "count"),
              completed=("overall_status", lambda x: (x=="COMPLETED").sum())
          )
    )
    summary["completion_rate"] = summary["completed"] / summary["total"]
    return summary
def sponsor_success_rate(df):
    summary = (
        df.groupby("lead_sponsor_class")
          .agg(
              total=("nct_id", "count"),
              completed=("overall_status", lambda x: (x=="COMPLETED").sum())
          )
    )
    summary["completion_rate"] = summary["completed"] / summary["total"]
    return summary
def results_delay_analysis(df):
    df2 = df.copy()
    df2["completion_date"] = pd.to_datetime(df2["completion_date"], errors="coerce")
    df2["first_post_date"] = pd.to_datetime(df2["first_post_date"], errors="coerce")
    df2["delay_days"] = (df2["first_post_date"] - df2["completion_date"]).dt.days
    return df2[df2["has_results"]==True]["delay_days"].describe()
def drug_repurposing_candidates(df):
    exploded = df.assign(
        intervention=df["intervention_names"].str.split(";"),
        condition=df["conditions"].str.split(";")
    ).explode("intervention").explode("condition")

    counts = exploded.groupby("intervention")["condition"].nunique()
    return counts.sort_values(ascending=False)
def phase_completion_by_phase(df):
    return (
        df.groupby("phases")
          .agg(total=("nct_id","count"),
               completed=("overall_status", lambda x: (x=="COMPLETED").sum()))
    )
def condition_trial_density(df):
    exploded = df.assign(condition=df["conditions"].str.split(";")).explode("condition")
    return exploded["condition"].value_counts()
def endpoint_trend_over_time(df):
    df2 = df.copy()
    df2["start_date"] = pd.to_datetime(df2["start_date"], errors="coerce")
    df2["year"] = df2["start_date"].dt.year
    return df2.groupby("year")["primary_outcome_measures"].count()
def masking_vs_completion(df):
    return (
        df.groupby("masking")
          .agg(total=("nct_id","count"),
               completed=("overall_status", lambda x: (x=="COMPLETED").sum()))
    )
def rare_disease_low_enrollment(df, threshold=50):
    return df[df["enrollment"] < threshold][["nct_id","conditions","enrollment"]]
def country_trend_over_time(df):
    df2 = df.copy()
    df2["start_date"] = pd.to_datetime(df2["start_date"], errors="coerce")
    df2["year"] = df2["start_date"].dt.year

    exploded = df2.assign(country=df2["countries"].str.split(";")).explode("country")
    return exploded.groupby(["year","country"]).size()
def sex_restriction_distribution(df):
    return df["sex"].value_counts()
def new_intervention_count_per_year(df):
    df2 = df.copy()
    df2["start_date"] = pd.to_datetime(df2["start_date"], errors="coerce")
    df2["year"] = df2["start_date"].dt.year

    exploded = df2.assign(intervention=df2["intervention_names"].str.split(";")).explode("intervention")
    return exploded.groupby("year")["intervention"].nunique()
def sponsor_trial_count(df):
    return df["lead_sponsor"].value_counts()
def terminated_trials(df):
    return df[df["overall_status"].isin(["TERMINATED","WITHDRAWN"])]
def enrollment_trend(df):
    df2 = df.copy()
    df2["start_date"] = pd.to_datetime(df2["start_date"], errors="coerce")
    df2["year"] = df2["start_date"].dt.year
    return df2.groupby("year")["enrollment"].mean()
def mesh_intervention_frequency(df):
    exploded = df.assign(mesh=df["mesh_interventions"].str.split(";")).explode("mesh")
    return exploded["mesh"].value_counts()
def phase_distribution(df):
    return df["phases"].value_counts()
def sponsor_condition_diversity(df):
    exploded = df.assign(condition=df["conditions"].str.split(";")).explode("condition")
    diversity = exploded.groupby("lead_sponsor")["condition"].nunique()
    return diversity.sort_values(ascending=False)
```

```python
import ijson
import os
import csv
from tqdm import tqdm

INPUT_FILE = "ctg-studies.json"

OUTPUT_DIR = "kg_output"
os.makedirs(OUTPUT_DIR, exist_ok=True)


def safe_get(d, *keys):
    for key in keys:
        if isinstance(d, dict):
            d = d.get(key)
        else:
            return None
    return d


def safe_list(value):
    return value if isinstance(value, list) else []


def open_writer(filename, fieldnames):
    f = open(os.path.join(OUTPUT_DIR, filename), "w", newline="", encoding="utf-8")
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    return f, writer


# -------------------------
# Open All Writers
# -------------------------

study_file, study_writer = open_writer("studies.csv", [
    "nct_id", "brief_title", "official_title",
    "overall_status", "start_date", "completion_date",
    "study_type", "phases", "enrollment",
    "allocation", "masking",
    "has_results"
])

arm_file, arm_writer = open_writer("arms.csv", [
    "nct_id", "arm_label", "arm_type", "arm_description"
])

intervention_file, intervention_writer = open_writer("interventions.csv", [
    "nct_id", "intervention_name", "intervention_type", "description"
])

condition_file, condition_writer = open_writer("conditions.csv", [
    "nct_id", "condition", "mesh_term"
])

sponsor_file, sponsor_writer = open_writer("sponsors.csv", [
    "nct_id", "lead_sponsor", "lead_sponsor_class"
])

location_file, location_writer = open_writer("locations.csv", [
    "nct_id", "facility", "city", "country", "lat", "lon"
])

outcome_file, outcome_writer = open_writer("outcomes.csv", [
    "nct_id", "measure", "timeframe", "type"
])

adverse_file, adverse_writer = open_writer("adverse_events.csv", [
    "nct_id", "event_term", "num_affected", "num_at_risk"
])

flow_file, flow_writer = open_writer("participant_flow.csv", [
    "nct_id", "period", "milestone", "group_id", "num_subjects"
])


# -------------------------
# Extraction Logic
# -------------------------

def extract_study(study):

    protocol = study.get("protocolSection", {})
    results = study.get("resultsSection", {})
    derived = study.get("derivedSection", {})

    nct_id = safe_get(protocol, "identificationModule", "nctId")

    # ---- STUDY TABLE ----
    study_writer.writerow({
        "nct_id": nct_id,
        "brief_title": safe_get(protocol, "identificationModule", "briefTitle"),
        "official_title": safe_get(protocol, "identificationModule", "officialTitle"),
        "overall_status": safe_get(protocol, "statusModule", "overallStatus"),
        "start_date": safe_get(protocol, "statusModule", "startDateStruct", "date"),
        "completion_date": safe_get(protocol, "statusModule", "completionDateStruct", "date"),
        "study_type": safe_get(protocol, "designModule", "studyType"),
        "phases": ";".join(safe_list(safe_get(protocol, "designModule", "phases"))),
        "enrollment": safe_get(protocol, "designModule", "enrollmentInfo", "count"),
        "allocation": safe_get(protocol, "designModule", "designInfo", "allocation"),
        "masking": safe_get(protocol, "designModule", "designInfo", "maskingInfo", "masking"),
        "has_results": study.get("hasResults")
    })

    # ---- ARMS ----
    for arm in safe_list(safe_get(protocol, "armsInterventionsModule", "armGroups")):
        arm_writer.writerow({
            "nct_id": nct_id,
            "arm_label": arm.get("label"),
            "arm_type": arm.get("type"),
            "arm_description": arm.get("description")
        })

    # ---- INTERVENTIONS ----
    for intervention in safe_list(safe_get(protocol, "armsInterventionsModule", "interventions")):
        intervention_writer.writerow({
            "nct_id": nct_id,
            "intervention_name": intervention.get("name"),
            "intervention_type": intervention.get("type"),
            "description": intervention.get("description")
        })

    # ---- CONDITIONS ----
    for condition in safe_list(safe_get(protocol, "conditionsModule", "conditions")):
        condition_writer.writerow({
            "nct_id": nct_id,
            "condition": condition,
            "mesh_term": None
        })

    for mesh in safe_list(safe_get(derived, "conditionBrowseModule", "meshes")):
        condition_writer.writerow({
            "nct_id": nct_id,
            "condition": None,
            "mesh_term": mesh.get("term")
        })

    # ---- SPONSOR ----
    sponsor_writer.writerow({
        "nct_id": nct_id,
        "lead_sponsor": safe_get(protocol, "sponsorCollaboratorsModule", "leadSponsor", "name"),
        "lead_sponsor_class": safe_get(protocol, "sponsorCollaboratorsModule", "leadSponsor", "class")
    })

    # ---- LOCATIONS ----
    for loc in safe_list(safe_get(protocol, "contactsLocationsModule", "locations")):
        location_writer.writerow({
            "nct_id": nct_id,
            "facility": loc.get("facility"),
            "city": loc.get("city"),
            "country": loc.get("country"),
            "lat": safe_get(loc, "geoPoint", "lat"),
            "lon": safe_get(loc, "geoPoint", "lon")
        })

    # ---- OUTCOMES ----
    for outcome in safe_list(safe_get(protocol, "outcomesModule", "primaryOutcomes")):
        outcome_writer.writerow({
            "nct_id": nct_id,
            "measure": outcome.get("measure"),
            "timeframe": outcome.get("timeFrame"),
            "type": "PRIMARY"
        })

    for outcome in safe_list(safe_get(protocol, "outcomesModule", "secondaryOutcomes")):
        outcome_writer.writerow({
            "nct_id": nct_id,
            "measure": outcome.get("measure"),
            "timeframe": outcome.get("timeFrame"),
            "type": "SECONDARY"
        })

    # ---- ADVERSE EVENTS ----
    for event in safe_list(safe_get(results, "adverseEventsModule", "otherEvents", "stats")):
        adverse_writer.writerow({
            "nct_id": nct_id,
            "event_term": None,
            "num_affected": event.get("numAffected"),
            "num_at_risk": event.get("numAtRisk")
        })

    # ---- PARTICIPANT FLOW ----
    for period in safe_list(safe_get(results, "participantFlowModule", "periods")):
        for milestone in safe_list(period.get("milestones")):
            for ach in safe_list(milestone.get("achievements")):
                flow_writer.writerow({
                    "nct_id": nct_id,
                    "period": period.get("title"),
                    "milestone": milestone.get("title"),
                    "group_id": ach.get("groupId"),
                    "num_subjects": ach.get("numSubjects")
                })


# -------------------------
# Streaming Runner
# -------------------------

def run():

    file_size = os.path.getsize(INPUT_FILE)
    total = 0

    with open(INPUT_FILE, "rb") as f:
        parser = ijson.items(f, "item")

        with tqdm(total=file_size,
                  unit="B",
                  unit_scale=True,
                  desc="Extracting KG Data") as pbar:

            for study in parser:
                extract_study(study)
                total += 1
                pbar.update(f.tell() - pbar.n)
                pbar.set_postfix(studies=total)

    print("Extraction complete.")
    print("Total studies processed:", total)


if __name__ == "__main__":
    run()
```

```python
import ijson
import os
import csv
from tqdm import tqdm


INPUT_FILE = "ctg-studies.json"
OUTPUT_FILE = "mesh_ancestors.csv"


def safe_get(d, *keys):
    for key in keys:
        if isinstance(d, dict):
            d = d.get(key)
        else:
            return None
    return d


def safe_list(x):
    return x if isinstance(x, list) else []


def extract_mesh_ancestors():

    file_size = os.path.getsize(INPUT_FILE)
    total = 0

    with open(OUTPUT_FILE, "w", newline="", encoding="utf-8") as out_csv:
        writer = csv.DictWriter(
            out_csv,
            fieldnames=[
                "nct_id",
                "mesh_id",
                "mesh_term",
                "ancestor_id",
                "ancestor_term"
            ]
        )
        writer.writeheader()

        with open(INPUT_FILE, "rb") as f:
            parser = ijson.items(f, "item")

            with tqdm(
                total=file_size,
                unit="B",
                unit_scale=True,
                desc="Extracting MeSH Ancestors"
            ) as pbar:

                for study in parser:

                    protocol = study.get("protocolSection", {})
                    derived = study.get("derivedSection", {})

                    nct_id = safe_get(protocol, "identificationModule", "nctId")

                    meshes = safe_list(
                        safe_get(derived, "conditionBrowseModule", "meshes")
                    )

                    ancestors = safe_list(
                        safe_get(derived, "conditionBrowseModule", "ancestors")
                    )

                    # Only write if both exist
                    if meshes and ancestors:
                        for mesh in meshes:
                            for ancestor in ancestors:
                                writer.writerow({
                                    "nct_id": nct_id,
                                    "mesh_id": mesh.get("id"),
                                    "mesh_term": mesh.get("term"),
                                    "ancestor_id": ancestor.get("id"),
                                    "ancestor_term": ancestor.get("term")
                                })

                    total += 1
                    pbar.update(f.tell() - pbar.n)
                    pbar.set_postfix(studies=total)

    print("\nExtraction complete.")
    print("Studies processed:", total)
    print("Output:", OUTPUT_FILE)


if __name__ == "__main__":
    extract_mesh_ancestors()
```

```python
import pandas as pd

# ---------------------------
# Load KG Tables
# ---------------------------

studies = pd.read_csv("kg_output/studies.csv")
conditions = pd.read_csv("kg_output/conditions.csv")
interventions = pd.read_csv("kg_output/interventions.csv")
sponsors = pd.read_csv("kg_output/sponsors.csv")


# ---------------------------
# Main Function
# ---------------------------

def competition_landscape(disease_keyword):

    # 1️⃣ Filter studies by disease
    cond_filtered = conditions[
        conditions["condition"].str.contains(disease_keyword, case=False, na=False)
    ]

    relevant_ncts = cond_filtered["nct_id"].unique()

    print(f"\nStudies found for '{disease_keyword}': {len(relevant_ncts)}")

    # 2️⃣ Subset relevant studies
    rel_studies = studies[studies["nct_id"].isin(relevant_ncts)]
    rel_sponsors = sponsors[sponsors["nct_id"].isin(relevant_ncts)]
    rel_interventions = interventions[interventions["nct_id"].isin(relevant_ncts)]

    # 3️⃣ Sponsor ranking
    sponsor_counts = (
        rel_sponsors.groupby("lead_sponsor")
        .size()
        .sort_values(ascending=False)
        .reset_index(name="num_trials")
    )

    # 4️⃣ Phase distribution
    phase_dist = (
        rel_studies.groupby("phases")
        .size()
        .reset_index(name="num_trials")
    )

    # 5️⃣ Intervention diversity per sponsor
    merged = rel_interventions.merge(rel_sponsors, on="nct_id")

    sponsor_diversity = (
        merged.groupby("lead_sponsor")["intervention_name"]
        .nunique()
        .reset_index(name="unique_interventions")
    )

    sponsor_summary = sponsor_counts.merge(
        sponsor_diversity,
        on="lead_sponsor",
        how="left"
    )

    sponsor_summary["crowding_score"] = (
        sponsor_summary["num_trials"] *
        sponsor_summary["unique_interventions"]
    )

    sponsor_summary = sponsor_summary.sort_values(
        by="crowding_score", ascending=False
    )

    print("\nTop Sponsors in Space:")
    print(sponsor_summary.head(10))

    print("\nPhase Distribution:")
    print(phase_dist)

    return sponsor_summary


# ---------------------------
# Example Run
# ---------------------------

summary = competition_landscape("Systemic Lupus Erythematosus")
import matplotlib.pyplot as plt

def plot_top_sponsors(summary):
    top = summary.head(10)
    plt.figure()
    plt.barh(top["lead_sponsor"], top["crowding_score"])
    plt.xlabel("Crowding Score")
    plt.title("Top Sponsors Competitive Density")
    plt.gca().invert_yaxis()
    plt.show()
plot_top_sponsors(summary)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

studies = pd.read_csv("kg_output/studies.csv")
sponsors = pd.read_csv("kg_output/sponsors.csv")

# Merge sponsor info
df = studies.merge(sponsors, on="nct_id", how="left")

# Define failure states
failure_states = ["TERMINATED", "WITHDRAWN", "SUSPENDED"]

df["failed"] = df["overall_status"].isin(failure_states).astype(int)
```

```python
heat_data = (
    df.groupby(["phases", "lead_sponsor_class"])
      .agg(
          total_trials=("nct_id", "count"),
          failures=("failed", "sum")
      )
      .reset_index()
)

heat_data["failure_rate"] = heat_data["failures"] / heat_data["total_trials"]

pivot_table = heat_data.pivot(
    index="phases",
    columns="lead_sponsor_class",
    values="failure_rate"
)
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(
    pivot_table,
    annot=True,
    fmt=".2f",
    cmap="Reds"
)
plt.title("Failure Risk Heatmap (Phase vs Sponsor Type)")
plt.ylabel("Phase")
plt.xlabel("Sponsor Class")
plt.show()
```

```python
import numpy as np

def drug_radar(drug_name):

    subset = merged[merged["intervention_name"] == drug_name]

    condition_counts = (
        subset["condition"]
        .value_counts()
        .head(8)
    )

    labels = condition_counts.index.tolist()
    values = condition_counts.values.tolist()

    angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
    values += values[:1]
    angles += angles[:1]

    fig, ax = plt.subplots(figsize=(6,6), subplot_kw=dict(polar=True))
    ax.plot(angles, values)
    ax.fill(angles, values, alpha=0.25)

    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(labels)

    plt.title(f"Repurposing Radar: {drug_name}")
    plt.show()
```
